```python
import pandas as pd

# Učitavanje podataka
data = pd.read_csv('surface temperature anomalies by year.csv')

# Prikaz prvih nekoliko redaka
data.head()

```

```python
# Prikaz svih naziva stupaca
print(data.columns)

# Provjeravamo je li potrebno ukloniti prvih nekoliko redaka
data = data.drop(index=0)

# Ponovno postavljanje naziva stupaca
data.columns = data.iloc[0]
data = data.drop(index=1)

# Provjeravanje prvih nekoliko redaka nakon čišćenja
print(data.head())
```

```python
# Pretvaranje svih vrijednosti u numeričke, ignoriranje pogrešaka
data = data.apply(pd.to_numeric, errors='coerce')

# Uklanjanje stupaca s previše nedostajućih vrijednosti
data = data.dropna(axis=1, how='any')

# Uklanjanje redaka s previše nedostajućih vrijednosti
data = data.dropna(axis=0, how='any')

# Provjeravanje prvih nekoliko redaka nakon čišćenja
print(data.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Pretvaranje natrag u DataFrame radi lakšeg rada
data_scaled = pd.DataFrame(data_scaled, columns=data.columns)

print(data_scaled.head())
```
